```python
def  Radioactive_decay_chain():
  t = 1000
  Pb = 0
  hol = [t]
  phol = [Pb]
  while t > 0:
          for i in range(t):
              if np.random.rand() < 1 - 2 ** (-1 / (3.052 * 60)):
                  t -= 1
                  Pb += 1
          hol.append(t)
          phol.append(Pb)
  t2 = np.arange(0, len(hol))
  plt.figure(figsize=(12, 12))
  plt.plot(t2, hol, label = "atoms" , color = "b",marker='*')
  plt.plot(t2, phol, label = "P atoms" , color = "r",marker='*')
  plt.xlabel("t")
  plt.ylabel("atom")
  plt.legend()
  plt.show()

```

```python
Radioactive_decay_chain()
```
