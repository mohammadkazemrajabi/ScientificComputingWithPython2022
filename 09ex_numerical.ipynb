 six-hump camelback function


```python
def function1(x): 

    return (4 - (2.1 * (x[0]**2)) + (x[0]**4)/3)*(x[0]**2) + x[0]*x[1] + (4*(x[1]**2)-4) * (x[1]**2)
result = minimize(function1, (0.25, 0.5))
result1= minimize(function1, (0,0))
```

```python
from scipy.integrate import odeint 
from scipy import fftpack
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def myfunc(y, t, list1):
        phi, omega = y
        g, r = list1
        return [omega, -(g/r)*np.sin(phi)]
def cal(g,r):  
  t =np.linspace(0, 200, 2000)
  results = odeint(myfunc, [20 , 0], t, args=([g, r],))
  plt.plot(t, results[:,0])
  plt.xlabel('t')
  plt.ylabel('a')
  plt.show()
cal(9.8,5)
```

```python
figure(figsize=(8, 6), dpi=80)
pic_fft= fft2(pic)
power = np.abs(pic_fft)
plt.imshow(power, norm=LogNorm(vmin=5))
plt.show()
```

```python
figure(figsize=(10, 8), dpi=80)
fft2(pic)[power > 2500] = 0
plt.imshow(ifft2(fft2(pic)).real)
```
